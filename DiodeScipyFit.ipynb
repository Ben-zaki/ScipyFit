```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

***Experiment carried on Diode 1N4007 Model*** 

```python
Vdmeas=np.array([0.20,0.3,0.39,0.47,0.52,0.55,0.57,0.6,0.62,0.67,0.69,0.71,0.73,0.74,
        0.75,0.76,0.77,0.78,0.79,0.80,0.81,0.82])
Idmeas=np.array([0.16*1e-3,1.7*1e-3,19.96*1e-3,0.11,0.32,0.59,0.90,1.89,2.46,
        5.48,8.5,11.75,14.98,18.15,21.40,24.60,27.86,31.25,34.46,
        41.00,47.60,50.01])
```

```python
e=1.6*1e-19
Tabs=300
k=1.38*1e-23
Vth=k*Tabs
Io=403*1e-12
```

```python
def DiodeModel(Vdmeas,n):
    return Io*(np.exp((e*Vdmeas)/(n*Vth))-1) 
```

```python
params, params_cov = curve_fit(DiodeModel,Vdmeas,Idmeas)
```

```python
print('Ideality Factor n is equal to :\n',round(params[0],3))
```

```python
plt.figure(figsize=(10,5))
plt.title("curve_fit_user_defined_funct\n IdealityFactor = "+str(round(params[0],3)),
          size=15,c='blue')
plt.scatter(Vdmeas,Idmeas,c = 'rebeccapurple',label='measured_data',s=80 )
plt.plot(Vdmeas,DiodeModel(Vdmeas, params[0]),'*',
         c ='chocolate',ms = 10.,label='fitted_data')
plt.xlabel('Vdiode',size=15,c='blue')
plt.ylabel('Idiode',size=15,c='blue')
plt.legend(loc='upper left',fontsize=12)
plt.grid(ls='--')
plt.savefig('IV_Diode1N4007.pdf')
plt.show();
```
